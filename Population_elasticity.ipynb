```python
import matplotlib.pylab as plt   # That gives plotting, and the next line makes plots appear inline in notebook
%matplotlib inline  
import numpy as np  # That gives numerical arrays and tools for manipulating them

import scipy.optimize as opt
import scipy.linalg as la
import math
```

```python
age_max = 50

A_mat = np.zeros([age_max+1,age_max+1])

A_mat[0,3:age_max+1] = 0.24


for i in np.arange(0,3):
    A_mat[i+1,i] = 0.0722**(1/3)

for i in np.arange(3,age_max):
    A_mat[i+1,i] = 0.942
```

Compute eigenvalues

```python
l,v = la.eig(A_mat)

#return largest eigenvalue
idx=np.argsort(np.abs(l))
l_sorted=l[idx]  
v_sorted=v[:,idx]   #rearrange eigenvectors in same order
lambda_max= l_sorted[-1]
print(lambda_max)
```

```python
l,v_left = la.eig(np.transpose(A_mat))

idx=np.argsort(np.abs(l))
l_sorted=l[idx]  
v_left_sorted=v_left[:,idx]   #rearrange eigenvectors in same order
v_left_max = v_left_sorted[:,-1]
```

```python
s = np.outer(v_left_max,v_max)/np.dot(v_left_max,v_max)

np.set_printoptions(precision=8,suppress=True)
# print(np.real(s))  #that real is just to nicely format the output here
```
